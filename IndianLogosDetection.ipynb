```python
def exit():
    global gui
    global root
    gui.quit()
    root.destroy()
```

```python
def save():
    if "saved_images" not in os.listdir():
        os.mkdir("saved_images")
    final_res.save("./saved_images/"+img_dis_path.split("/")[-1])
```

```python
img_dis_path = 'white.png'
model_path = ''
i=1
def file_opener():
    global file_path
    global img_dis_path
    global i
    inp_file = filedialog.askopenfilename()
    img_dis_path = inp_file
    
    img_new=Image.open(img_dis_path)
    img_new = img_new.resize((600, 500), Image.ANTIALIAS)
    img2 = ImageTk.PhotoImage(img_new)
    image_display.configure(image=img2)
    image_display.image = img2

def browseFiles():
    global model_path
    model_path = filedialog.askopenfilename(initialdir = "/",
                                          title = "Select a File",
                                          filetypes = (("YOLOv5 Weights",
                                                        "*.pt*"),
                                                       ("all files",
                                                        "*.*")))    
saved_path = ''
def detect():
    global fin_img
    global i
    global saved_path
    global model_path
    global iou_slider
    global final_res
    print(img_dis_path)
    #new_img_dis_path =  "\"{}\"".format(img_dis_path)
    model = torch.hub.load('ultralytics/yolov5', 'custom', path=model_path)
    model.conf = float(iou_slider.get())/100
    res = model(img_dis_path)
    final_res = np.squeeze(res.render())
    final_res = Image.fromarray(final_res,"RGB")
    fin_img = final_res.resize((600,500), Image.ANTIALIAS)
    fin_img = ImageTk.PhotoImage(fin_img)
    #image_display.clear()
    image_display.configure(image=fin_img)
    image_display.image = fin_img
    
    #i+=1
    #!python ./yolov5-master/detect.py --weights best.pt --img 448 --conf 0.2 --source $new_img_dis_path
    #saved_path = "./yolov5-master/runs/detect/exp"+str(i)+ "/" + new_img_dis_path.split("/")[-1][:-1]
    #fin_img = Image.open(saved_path)
    #fin_img = fin_img.resize((600,500), Image.ANTIALIAS)
    #fin_img = ImageTk.PhotoImage(fin_img)
    #image_display.clear()
    #image_display.configure(image=fin_img)
    #image_display.image = fin_img
```

```python
root = Tk()
root.withdraw()
gui = Toplevel()
gui.config(background =  "Black") 
gui.title("Logo Detection")  
gui.geometry("1280x720")

logo_image = Image.open("LogoTitle.png")
logo_image = ImageTk.PhotoImage(logo_image.resize((250, 125), Image.ANTIALIAS))
#logo_header = Label(gui,text="LOGO DETECTION",fg="white",bg="black",font=("cambria",25),borderwidth=2)
logo_header = Label(gui,image=logo_image,bg="black")
logo_header.place(x=500,y = 0)

image = Image.open(img_dis_path)
image = image.resize((600, 500), Image.ANTIALIAS)
my_img = ImageTk.PhotoImage(image)
image_display = Label(gui,image=my_img,borderwidth=5)
image_display.place(x=325,y=110)

save_image = PhotoImage(file="save.png")
save_button = Button(gui,image=save_image,text="Save",fg="black",bg="black",borderwidth=0,command=save)
save_button.place(x=175,y=325)

browse_image = PhotoImage(file="browse.png")
browse_button = Button(gui,image=browse_image,text="Browse\nImage",fg="black",bg="black",borderwidth=0,command=file_opener)
browse_button.place(x=175,y=475)

model_image = PhotoImage(file="model.png")
model_button = Button(gui,image=model_image,text="Load\nModel",fg="black",bg="black",borderwidth=0,command=browseFiles)
model_button.place(x=175,y=175)

detect_image = PhotoImage(file="detect.png")
detect_button = Button(gui,image=detect_image,text="Detect\nImage",fg="black",bg="black",borderwidth=0,command=detect)
detect_button.place(x=975,y=325)

iou_label = Label(gui,text="Confidence\nThreshold:",font="Cambria",bg="black",fg="white")
iou_label.place(x=975,y=107)

iou_slider = Scale(gui,from_=0,to=100,bg="black",fg="white",resolution=5)
iou_slider.place(x=1000,y=180)

exit_image = PhotoImage(file="exit.png")
exit_button = Button(gui,image=exit_image,text="Exit",fg="black",bg="black",borderwidth=0,command=exit)
exit_button.place(x=975,y=475)
mainloop()
```
